```python
import pandas as pd
import numpy as np
```

```python
import requests
from bs4 import BeautifulSoup
```

```python
# Identify the URL

url = 'https://www.skullcandy.in/product-category/best-sellers/?gad_source=1&gclid=CjwKCAiAtNK8BhBBEiwA8wVt9wZTu1UvVoIhfXcuqheo70CmyVJTDo6NRwL8a32MHGdDHl2nCG94rhoCOQwQAvD_BwE'
```

```python
# Loading the WebPage in Memory using requests library

page = requests.get(url)
```

```python
# Check the Status Code of the Page

page.status_code
```

```python
# Extracting the HTML Code of the WebPage

htmlCode = page.text
soup = BeautifulSoup(htmlCode )
```

```python
htmlCode
```

```python
content = soup.find_all('div', attrs = {'class': 'product_detail'})
print(content)
```

```python
text =''
for i in content:
  text += i.text
print(text)
```

```python
data=text.split('\n')
print(data)
```

```python
maindata=[]
for i in data:
  x=i.strip().split('\n')
  if x!=['']:
    maindata.append(x)
print(maindata)
```

```python
names=maindata[::2]
price=maindata[1::2]
product=[]
price_list=[]
for i in names:
  product.append(i[0])
for i in price:
  price_list.append(i[0])
print(product)
print(price_list)
```

# **Creating a DataFrame and saveing it in CSV file**

```python
df = pd.DataFrame({'Product_names' : product,'Product_prices' : price_list})
```
